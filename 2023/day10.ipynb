```python
import numpy as np
from collections import Counter

input_text = open('inputs/day10.txt', 'r').readlines()
input_array = np.array([list(line.strip()) for line in input_text])
input_array = np.pad(input_array, 1)
zero_array = np.zeros(input_array.shape)
```

```python
def get_start_coords(arr):
    itemindex = np.where(arr == "S")
    return itemindex[0][0], itemindex[1][0], 

def find_connections_to_start(start_coords):
    connections = []
    y, x = start_coords
    if input_array[y-1, x] in ["|", "F", "7"]:
        connections.append(((y-1, x), "up"))
    if input_array[y, x+1] in ["-", "J", "7"]:
        connections.append(((y, x+1), "right"))
    if input_array[y+1, x] in ["|", "J", "L"]:
        connections.append(((y+1, x), "down"))
    if input_array[y, x-1] in ["-", "F", "L"]:
        connections.append(((y, x-1), "left"))

    return connections     

def follow_route(current_position, direction):
    current_pipe = input_array[current_position]
    count = 1
    while current_pipe != "S":
        save_original_pos = current_position
        if current_pipe == "|":
            if direction == "up":
                current_position = (current_position[0]-1, current_position[1])
            elif direction == "down":
                current_position = (current_position[0]+1, current_position[1])

        elif current_pipe == "-":
            if direction == "left":
                current_position = (current_position[0], current_position[1]-1)
            elif direction == "right":
                current_position = (current_position[0], current_position[1]+1)

        elif current_pipe == "L":
            if direction == "left":
                current_position = (current_position[0]-1, current_position[1])
                direction = "up"
            elif direction == "down":
                current_position = (current_position[0], current_position[1]+1)
                direction = "right"

        elif current_pipe == "J":
            if direction == "right":
                current_position = (current_position[0]-1, current_position[1])
                direction = "up"
            elif direction == "down":
                current_position = (current_position[0], current_position[1]-1)
                direction = "left"

        elif current_pipe == "7":
            if direction == "right":
                current_position = (current_position[0]+1, current_position[1])
                direction = "down"
            elif direction == "up":
                current_position = (current_position[0], current_position[1]-1)
                direction = "left"
            
        elif current_pipe == "F":
            if direction == "left":
                current_position = (current_position[0]+1, current_position[1])
                direction = "down"
            elif direction == "up":
                current_position = (current_position[0], current_position[1]+1)
                direction = "right"

        current_pipe = input_array[current_position]
        count += 1

        # Mark route for part 2
        zero_array[save_original_pos] = 1
        
    return int(count / 2)

def remove_non_used():
    for y in range(len(input_array)):
        for x in range(len(input_array[0])):
            if (zero_array[y, x] == 0) and (input_array[y, x] != "S"):
                input_array[y, x] = "."

def count_inside():
    inside_count = 0
    for line in input_array:
        for idx in range(len(line)):
            if line[idx] == ".":
                left_bars = line[:idx]
                if (
                    Counter(left_bars)["|"] +
                    Counter(left_bars)["L"] +
                    Counter(left_bars)["J"]
                    + Counter(left_bars)["S"] 
                    ) % 2 != 0: 
                    inside_count += 1
    return inside_count

start_coords = get_start_coords(input_array)
start_connections = find_connections_to_start(start_coords)
print(f"Part 1 Answer: {follow_route(*start_connections[0])}")
remove_non_used()
print(f"Part 2: Answer: {count_inside()}")
```
