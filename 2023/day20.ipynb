```python
import math

def process_input(text):
    config = {}
    for line in text:
        start, end = line.split("->")
        start = start.strip()
        end = end.strip()
        end = [next.strip() for next in end.split(",")]
        if start == "broadcaster":
            config[start] = {"outputs":end}
        elif start[0] == "%":
            config[start[1:]] = {
                "type" : "%",
                "status" : "off",
                "outputs": end
            }
        elif start[0] == "&":
            config[start[1:]] = {
                "type" : "&",
                "outputs": end
            }

    for module in config:
        if (module != "broadcaster") and (config[module]["type"] == "&"):
            memory = {}
            for module2 in config:
                for output in config[module2]["outputs"]:
                    if module == output:
                        memory[module2] = "low"
            config[module]["memory"] = memory

    return config

def send_pulses(module, pulse):
    for output in configuation[module]['outputs']:
        next_module.append((output, pulse, module))

input_text = open("inputs/day20.txt", "r").readlines()
configuation = process_input(input_text)
low_pulses = 0
high_pulses = 0
button_press = 0
cycles = {}
finished = False

while not finished:
    first_pulse = "low"
    first_round = "button"
    next_module = [("broadcaster", first_pulse, first_round)]

    for module, pulse, pulse_from in next_module:

        if module == "broadcaster":
            send_pulses(module, pulse)

        elif module == "rx":
            pass

        elif configuation[module]["type"] == "%":
            if pulse == "low":
                if configuation[module]["status"] == "off":
                    configuation[module]["status"] = "on"
                    send_pulses(module, "high")
                else:
                    configuation[module]["status"] = "off"
                    send_pulses(module, "low")

        elif configuation[module]["type"] == "&":
            configuation[module]["memory"][pulse_from] = pulse
            if set(configuation[module]['memory'].values()) == set(["high"]):
                send_pulses(module, "low")   
            else:
                send_pulses(module, "high")  

        if (module == "kj") and (pulse == "high"):
            if pulse_from not in cycles.keys():
                cycles[pulse_from] = button_press + 1
            if len(cycles.keys()) == 4:
                finished = True

    if button_press <= 1000:
        high_pulses += len([1 for _, pulse, _ in next_module if pulse == "high"])
        low_pulses += len([1 for _, pulse, _ in next_module if pulse == "low"])
    button_press += 1

print(f"Part 1 Answer: {high_pulses * low_pulses}")
print(f"Part 2 Answer: {math.prod(cycles.values())}")
```
