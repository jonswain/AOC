```python
import numpy as np


def parse_input(input_text):
    map_list = [list(line) for line in input_text.split("\n")]
    map_array = np.array(map_list)
    start = np.where(map_array == "^")
    start_coords = (start[0][0], start[1][0])
    return map_array, start_coords


def get_next_position(coords, direction):
    if direction == "up":
        if coords[0] - 1 < 0:
            raise IndexError
        return (coords[0] - 1, coords[1])
    elif direction == "down":
        return (coords[0] + 1, coords[1])
    elif direction == "left":
        if coords[1] - 1 < 0:
            raise IndexError
        return (coords[0], coords[1] - 1)
    elif direction == "right":
        return (coords[0], coords[1] + 1)


def turn_right(direction):
    if direction == "up":
        return "right"
    elif direction == "down":
        return "left"
    elif direction == "left":
        return "up"
    elif direction == "right":
        return "down"


def explore_map(map_array, start_coords, guard_direction, part=1):
    guard_coords = start_coords
    coords_log = set()
    while True:
        try:
            coords_log.add((guard_coords, guard_direction))
            next_coords = get_next_position(guard_coords, guard_direction)
            while map_array[next_coords[0], next_coords[1]] == "#":
                guard_direction = turn_right(guard_direction)
                next_coords = get_next_position(guard_coords, guard_direction)
            guard_coords = next_coords
            if (part == 2) and ((guard_coords, guard_direction) in coords_log):
                coords_log.add("stuck in loop")
                return coords_log
        except IndexError:
            return coords_log


input_text = open("inputs/day6.txt", "r").read()
map_array, start_coords = parse_input(input_text)
guard_direction = "up"

# Part 1
coords_log = explore_map(map_array, start_coords, guard_direction)
positions = set([x[0] for x in coords_log])

# Part 2
count = 0
for i, j in list(positions):
    if map_array[i, j] == "^":
        continue

    new_map, start_coords = parse_input(input_text)
    guard_coords = start_coords
    guard_direction = "up"

    new_map[i, j] = "#"

    route = explore_map(new_map, start_coords, guard_direction, part=2)
    if "stuck in loop" in route:
        count += 1

# Answers
print(f"Part 1: {len(positions)}")
print(f"Part 2: {count}")
```
